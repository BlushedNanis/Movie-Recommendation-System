# Popularity-Based Filtering

## Load the data

```
import pandas as pd

movies_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
credits_df = pd.read_csv("credits.csv")
```

## Calculate weighted rating

### Formula

WR = (v / (v+m)) x R + (m / (v+m)) x C

- v - number of votes for a movie (already in movies_df)

- m - minimum number of votes required

- R - average rating of the movie (already in movies_df)

- C - average rating across all movies

### Calculate missing variables

```
m = movies_df['vote_count'].quantile(0.5)
m
```

```
c = movies_df['vote_average'].mean()
c
```

### Filter movies by vote count

```
filtered_movies_df = movies_df.loc[movies_df['vote_count'] >= m]
filtered_movies_df.head()
```

### Function to calculate weighted rating

```
def weighted_rating(df, m=m, c=c):
    v = df['vote_count']
    R = df['vote_average']
    wr = ((v/(v+m)) * R) + (m/(m+v) * c)
    return wr
```

### Add column to df with the weighted rating

```
filtered_movies_df["weighted_rating"] = filtered_movies_df.apply(weighted_rating, axis=1)
filtered_movies_df.head()
```

## Popularity-based output

### Sort data frame by weighted rating

```
filtered_movies_df = filtered_movies_df.sort_values('weighted_rating', ascending=False)
filtered_movies_df[['title', 'vote_count', 'vote_average', 'weighted_rating']].head(10)
```

### Dictionary output

```
filtered_movies_df[['title','weighted_rating']].to_dict('records')
```
